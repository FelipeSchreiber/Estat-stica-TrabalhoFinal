```python
import pandas as pd
import unicodedata
data = pd.read_csv('dadosMedicos.csv',delim_whitespace=True, usecols =[0,1,2,3], header=0)
data
```

# Histograma e Função Distribuição Empírica

## Idade

```python
idade = data["IDADE"]
desvioPadrao = idade.std()
binwidth = 3.49*desvioPadrao*(math.pow(idade.count(),(-0.333333333)))
plt.title('Histograma idade')
plt.xlabel('idade')
plt.ylabel('frequencia')
plt.hist(idade, bins=np.arange(min(idade), max(idade) + binwidth, binwidth),color=['orange'])
plt.savefig('Histograma idade')
plt.show()

sns.distplot(idade,  kde = True, bins=np.arange(min(idade), max(idade) + binwidth, binwidth))
plt.title('PDF idade')
plt.xlabel('idade')
plt.ylabel('densidade de probabilidade')
plt.savefig('PDF idade')
plt.show()

x = np.sort(idade)
y = np.arange(1, len(x)+1)/float (len(x))
plt.title('CDF idade')
plt.xlabel('idade')
plt.ylabel('P[X =< x]')
plt.plot(x, y, marker='.',linestyle='none')
plt.margins(0.02)
plt.savefig('CDF idade')
plt.show()
```

# Parametrizando distribuições

## Idade

```python
interval = np.linspace(idade.min(),idade.max(),50000) #delimita o intervalo da plotagem
exp_lambda = 1/idade.mean() #estima o parametro lambda analiticamente
exp= exp_lambda* np.exp(-(exp_lambda*interval))#gera o gráfico
plt.plot(interval,exp,label ="Exponencial")#plota
print "Valor do ",unicodedata.lookup("GREEK SMALL LETTER LAMDA")," da Exponencial: ", exp_lambda, "\n"


mi = idade.mean()
sigma = idade.std()
normal = (1 / (np.sqrt(2 * np.pi * np.power(sigma, 2))))*(np.power(np.e, -(np.power((interval - mi), 2) / (2 * np.power(sigma, 2)))))
plt.plot(interval, normal, label="Normal")
print "Valor de ", unicodedata.lookup("GREEK SMALL LETTER MU"), " (média) da Normal: ", mi
print "Valor de ", unicodedata.lookup("GREEK SMALL LETTER SIGMA"), " (desvio padrão) da Normal: ", sigma, "\n"
    
mi = np.sum(np.log(idade))/idade.count() #aplica o log natural em cada uma das amostras, faz o somatório 
#e divide pelo número total de amostras para determinar o parametro scale da lognormal
sigmaAoQuadrado = np.sum((np.log(idade)-mi)**2)/idade.count() #o parametro shape da lognormal é obtido
#fazendo o desvio padrão com relação à média mi - parametro scale obtido anteriormente
lognormal= (np.exp(-(np.log(interval) - mi)**2 / (2 * sigmaAoQuadrado))/ (interval * (np.sqrt(sigmaAoQuadrado)) * np.sqrt(2 * np.pi)))
plt.plot(interval, lognormal, label="Lognormal")
print "Valor de ", unicodedata.lookup("GREEK SMALL LETTER MU"), " (scale) da lognormal: ", mi
print "Valor de ", unicodedata.lookup("GREEK SMALL LETTER SIGMA"),"^2 (shape) da lognormal: ",sigmaAoQuadrado, "\n"
                             
shape, loc, scale = stats.weibull_min.fit(idade,floc=0)#obtem os parametros da weibull que maximizam a Likelihood
weibull = stats.weibull_min.pdf(interval, shape, loc=0, scale=scale)
plt.plot(interval, weibull, label="Weibull")
print "Valor de ", unicodedata.lookup("GREEK SMALL LETTER BETA")," (scale) da weibull: ",scale
print "Valor de ", unicodedata.lookup("GREEK SMALL LETTER ALPHA")," (shape) da weibull: ",shape 
print "Valor de ", unicodedata.lookup("GREEK SMALL LETTER GAMMA")," (location) da weibull: ",loc,"\n"

fit_alpha, fit_loc, fit_beta=stats.gamma.fit(idade) #obtem os parametros da gamma que maximizam a Likelihood
y1 = stats.gamma.pdf(interval, a=fit_alpha, scale=fit_beta, loc = fit_loc)
plt.plot(interval, y1, label="gamma")
print "Valor de ", unicodedata.lookup("GREEK SMALL LETTER BETA"),"  da gamma: ",fit_beta
print "Valor de ", unicodedata.lookup("GREEK SMALL LETTER ALPHA"),"  da gamma: ",fit_alpha
print "Valor do location da gamma: ",fit_loc,"\n"

sns.distplot(idade,  kde = True, hist=False,label='empirica')
plt.title('Distribuicoes idade')
plt.xlabel('idade')
plt.ylabel('distribuicao')
plt.legend(loc='upper right')
plt.savefig('distribuicoesIdade')
plt.show()
```

# Inferência Baysiana

```python
probabilidade_do_intervalo=[]
binwidth=(cargaFinal.max()-cargaFinal.min())/float(20)
limite_inferior_do_intervalo=30.00
limite_superior_do_intervalo=30.00+binwidth
hipotese_intervalo=[]
vo2_menor_condicionada_no_intervalo =[]
vo2_maior_condicionada_no_intervalo =[]

#primeiramente vamos dividir a carga final em 20 intervalos iguais. Escolheu-se esse valor pois no histograma obtido 
#no 1 item tinhamos 16 bins e queriamos um multiplo de 4 mais proximo de forma a obter intevalos bem definidos ao inves 
#de dízimas periodicas
for i in range(0,19):
    #utilizou-se a funcao round apenas para fins de exibicao
    #nao aparecer muitos zeros
    hipotese_intervalo.append((round(limite_inferior_do_intervalo,4),round(limite_superior_do_intervalo,4)))
    #aqui a probabilidade_do_intervalo será a probabilidade de um certo individuo possuir carga final dentro de um intervalo
    #como, nao sabemos o valor exato de antemao, contaremos a frequencia dos valores que se situam nesse intervalo 
    #para obter a pmf da probabilidade_do_intervalo de forma empírica, dividindo depois a frequencia pela quantidade total de amostras
    frequencia_do_intervalo=data[(cargaFinal>=limite_inferior_do_intervalo) & (cargaFinal<limite_superior_do_intervalo)]
    probabilidade_do_intervalo.append(frequencia_do_intervalo["Carga Final"].count()/float(cargaFinal.count()))
    #Para obter a condicional, isto é, a probabilidade da VO2 ser menor do que 35 dado que
    #a carga final está naquele intervalo contaremos quantos individuos se enquadram, ao mesmo tempo, nesses dois
    #critérios. Então dividiremos pelo espaço amostral, isto é, a frequencia de individuos que possui a carga final
    #nesse intervalo. Em outras palavras:
    #Seja A o conjunto de individuos que possuem VO2 < 35
    #Seja B o conjunto de individuos que possuem carga final nesse intervalo
    #Entao P(A | B)= P(A,B)/P(B)
    #Mas P(A,B)/P(B) = Numero total de individuos que estao na intersecao de A e B/Numero de individuos do conjunto B 
    frequencia_vo2_menor_condicionada_no_intervalo =data[(data["Carga Final"]>=limite_inferior_do_intervalo) & (data["Carga Final"]<limite_superior_do_intervalo) & (data["VO2"]<35.0)]
    vo2_menor_condicionada_no_intervalo.append(frequencia_vo2_menor_condicionada_no_intervalo["VO2"].count()/float(frequencia_do_intervalo["Carga Final"].count()))
    #De maneira analoga, para o conjunto C tal que VO2 >= 35 
    frequencia_vo2_maior_condicionada_no_intervalo =data[(data["Carga Final"]>=limite_inferior_do_intervalo) & (data["Carga Final"]<limite_superior_do_intervalo) & (data["VO2"]>=35.0)]
    vo2_maior_condicionada_no_intervalo.append(frequencia_vo2_maior_condicionada_no_intervalo["VO2"].count()/float(frequencia_do_intervalo["Carga Final"].count()))
    #Agora temos que passar para o proximo intervalo, incrementando o valor do "binwidth"(largura do intervalo)
    limite_inferior_do_intervalo+= binwidth 
    limite_superior_do_intervalo+= binwidth

hipotese_intervalo.append((round(limite_inferior_do_intervalo,1),round(limite_superior_do_intervalo,1)))

frequencia_do_intervalo=data[(data["Carga Final"]>=limite_inferior_do_intervalo) & (data["Carga Final"]<=limite_superior_do_intervalo)]
probabilidade_do_intervalo.append(frequencia_do_intervalo["Carga Final"].count()/float(data["Carga Final"].count()))

frequencia_vo2_menor_condicionada_no_intervalo =data[(data["Carga Final"]>=limite_inferior_do_intervalo) & (data["Carga Final"]<=limite_superior_do_intervalo) & (data["VO2"]<35.0)]
vo2_menor_condicionada_no_intervalo.append(frequencia_vo2_menor_condicionada_no_intervalo["VO2"].count()/float(frequencia_do_intervalo["Carga Final"].count()))
    
frequencia_vo2_maior_condicionada_no_intervalo =data[(data["Carga Final"]>=limite_inferior_do_intervalo) & (data["Carga Final"]<=limite_superior_do_intervalo) & (data["VO2"]>=35.0)]
vo2_maior_condicionada_no_intervalo.append(frequencia_vo2_maior_condicionada_no_intervalo["VO2"].count()/float(frequencia_do_intervalo["Carga Final"].count()))
    

bayes_numerador_menor_35=[]
bayes_numerador_maior_35=[]
for i in range(0,20):
    #O numerador será P(A|B)*P(B)
    bayes_numerador_menor_35.append(probabilidade_do_intervalo[i]*vo2_menor_condicionada_no_intervalo[i])
    bayes_numerador_maior_35.append(probabilidade_do_intervalo[i]*vo2_maior_condicionada_no_intervalo[i])

# Para achar o P(A), isto é, P(VO2<35), aplicaremos o Teorema da Probabilidade Total, onde:
#P(A)= Somatório de P(A|B)*P(B), para todo B
total_menos_35=np.sum(bayes_numerador_menor_35)
#print "totalMenos: ",total_menos_35
#De maneira análoga, para o P(C) = P(VO2>=35)
total_mais_35=np.sum(bayes_numerador_maior_35)

posterior_menos_35=[]
posterior_mais_35 =[]

for i in range(0,20):
    #Agora basta dividirmos o numerador de Bayes pelo P(A)
    posterior_menos_35.append(bayes_numerador_menor_35[i]/float(total_menos_35))
    #Da mesma forma, dividiremos o numerador de Bayes para VO2>=35 pelo P(C)
    posterior_mais_35.append(bayes_numerador_maior_35[i]/float(total_mais_35))

predict=[]    
for i in range(0,20):
    predict.append(posterior_menos_35[i]*vo2_maior_condicionada_no_intervalo[i])

prob=np.sum(predict)

inference_menos_35= pd.DataFrame({"hipotese":hipotese_intervalo,
                              "prior":probabilidade_do_intervalo,
                              "likelihood(<35)" : vo2_menor_condicionada_no_intervalo,
                              "Bayes Num":bayes_numerador_menor_35,
                              "Posterior":posterior_menos_35})

inference_mais_35= pd.DataFrame({"hipotese":hipotese_intervalo,
                              "prior":probabilidade_do_intervalo,
                              "likelihood(>=35)" :vo2_maior_condicionada_no_intervalo,
                              "Bayes Num":bayes_numerador_maior_35,
                              "Posterior":posterior_mais_35})

inference_cond= pd.DataFrame({"hipotese":hipotese_intervalo,
                              "Posterior 1": posterior_menos_35,
                              "lk(<35)" :vo2_menor_condicionada_no_intervalo,
                              "Bayes Num1":bayes_numerador_maior_35,
                              "prior":probabilidade_do_intervalo,
                              "Posterior 1": posterior_menos_35,
                              "lk(>=35)" : vo2_maior_condicionada_no_intervalo,
                              "predict": predict})

print "hipotese VO2 max < 35\n\n",inference_menos_35 ,"\n"
print total_menos_35,"\n\n\n\n"
print "hipotese VO2 max >= 35\n\n",inference_mais_35,"\n"
print total_mais_35,"\n\n\n\n"
print "hipotese VO2 max >= 35|VO2 max < 35 anteriormente \n\n"
inference_cond
print total_menos_35
print "prob: ", prob,"\n\n\n\n"
```
